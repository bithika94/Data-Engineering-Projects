Stock Price Data

```python
import pandas as pd
import os

stock_prices ={}

for fn in os.listdir("prices"):
    name = fn.split(".")[0]
    stock_prices[name] = pd.read_csv(os.path.join("prices", fn))

```

```python
stock_prices["aapl"].head()
```

Calculate the average closing price of a stock

```python
Stock_closing_price = {}
for name in stock_prices:
    Stock_closing_price[name] = stock_prices[name]["close"].mean()
```

Displaying the average closing prices

```python
for name in stock_prices:
    print(name, stock_prices[name]["close"])
```

Minimum & Maximum Closing Price

```python
pairs = [(Stock_closing_price[name], name) for name in Stock_closing_price]
pairs.sort()

#Two min avg
print("Two minimum average closing prices:")
print(pairs[0])
print(pairs[1])

#Two max avg
print("Two maximum average closing prices:")
print(pairs[-1])
print(pairs[-2])
```

Organizing the Trades Per Day

```python
Stock_Volume = {}
for stock in stock_prices:
    for index, row in stock_prices[stock].iterrows():
        volume = row["volume"]
        day = row["date"]
        pair = (volume, stock)
        if day not in Stock_Volume:
            Stock_Volume[day] = []
        Stock_Volume[day].append(pair)
    
```

Finding the Most Traded Stock Each Day

```python
most_traded_by_day = {}
for day in Stock_Volume:
    Stock_Volume[day].sort()
    most_traded_by_day[day] = Stock_Volume[day][-1]
```

```python
#Testing for few dates
print(most_traded_by_day['2007-01-30'])
print(most_traded_by_day['2007-02-01'])
```

High Volume Days

```python
daily_volume = []
for day in Stock_Volume:
    day_volume = sum([volume for volume, _ in Stock_Volume[day]])
    daily_volume.append((day_volume, day))

daily_volume.sort()
daily_volume[-10:]
```

Finding Profitable Stocks

```python
percentages = []
for stock in stock_prices:
    prices = stock_prices[stock]
    initial = prices.loc[0,"close"]
    final = prices.loc[prices.shape[0]-1,"close"]
    percentage = 100 * (final-initial)/initial
    percentages.append((percentage, stock))

percentages.sort()
percentages[-10:]
    
```
