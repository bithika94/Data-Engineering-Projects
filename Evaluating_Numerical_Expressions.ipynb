```python
from linked_list import LinkedList
class Stack(LinkedList):
    def push(self, data):
        self.append(data)

    def peek(self):
        return self.tail.data

    def pop(self):
        ret = self.tail.data
        if self.length == 1:
            self.tail = self.head = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        self.length -= 1
        return ret
```

Tokenize a function

```python
def tokenize(expression):
    return expression.split()
```

```python
print(tokenize("12 2 4 + / 21 *"))
```

Processing an Operator

```python
def process_minus(stack):
    top = stack.pop()
    second_to_top = stack.pop()
    result = second_to_top - top
    stack.push(result)

def process_plus(stack):
    top = stack.pop()
    second_to_top = stack.pop()
    result = second_to_top + top
    stack.push(result)

def process_times(stack):
    top = stack.pop()
    second_to_top = stack.pop()
    result = second_to_top * top
    stack.push(result)

def process_divide(stack):
    top = stack.pop()
    second_to_top = stack.pop()
    result = second_to_top / top
    stack.push(result)

def process_pow(stack):
    top = stack.pop()
    second_to_top = stack.pop()
    result = second_to_top ** top
    stack.push(result)
    
```

```python
def evaluate_postfix(expression):
    tokens = tokenize(expression)
    my_stack = Stack()
    for token in tokens:
        if token == "+":
            process_plus(my_stack)
        elif token == "-":
            process_minus(my_stack)
        elif token == "*":
            process_times(my_stack)
        elif token == "/":
            process_divide(my_stack)
        elif token == "**":
            process_pow(my_stack)
        else:
            # The token is a number, push it to the stack as float
            my_stack.push(float(token))
    # After processing all tokens, return the result
    return my_stack.pop()
```

```python
expression = [
    "4 6 -",
    "4 1 2 9 3 / * + 5 - *",
    "1 2 + 3 -",
    "1 2 - 3 +",
    "10 3 5 * 16 4 - / +",
    "5 3 4 2 - ** *",
    "12 2 4 + / 21 *",
    "1 1 + 2 **",
    "1 1 2 ** +"
]

for exp in expression:
    print(evaluate_postfix(exp))
```

Precedence dictionary

```python
precendence = {
    "+": 1,
    "-": 1,
    "*": 2,
    "/": 2,
    "**": 3
}

print(precendence["/"] < precendence["-"])
print(precendence["+"] < precendence["*"])
print(precendence["+"] < precendence["-"])
print(precendence["/"] < precendence["**"])
```

Processing tokens in infix to postfix conversions

```python
def process_opening_parenthesis(stack):
    stack.push("(")
    
def process_closing_parenthesis(stack, postfix):
    # Add tokens until we find the open bracket
    while stack.peek() != "(":
        postfix.append(stack.pop())
    # Remove the opening bracket
    stack.pop()
        
```

Handling Operators

```python
def process_operator(stack, postfix, operator):
    while len(stack)>0 and stack.peek() in precedence and precendence[stack.peek()] >= precedence[operator]:
        postfix.append(stack.pop())
    stack.push(operator)    

    
```

```python
def process_number(postfix,number):
    postfix.append(number)
```

Implementing the Shunting-yard Algorithm

```python
def infix_to_postfix(expression):
    tokens = tokenize(expression)
    stack = Stack()
    postfix = []
    for token in tokens:
        if token == "(":
            process_opening_parenthesis(stack)
        elif token == ")":
            process_closing_parenthesis(stack, postfix)
        elif token in precedence:
            process_operator(stack, postfix, token)
        else:
            process_number(postfix, token)
    while len(stack) > 0:
        postfix.append(stack.pop())
    return " ".join(postfix)
```

Evaluating Infix Expressions

```python
def evaluate(expression):
    postfix_expression = infix_to_postfix(expression)
    return evaluate_postfix(postfix_expression)
```

```python
expressions = [
    "1 + 1",
    "1 * ( 2 - ( 1 + 1 ) )",
    "4 * ( 1 + 2 * ( 9 / 3 ) - 5 )",
    "10 + 3 * 5 / ( 16 - 4 * 1 )",
    "2 * 2 * 2 * 2 * 2 * 2 * 2 * 2",
    "2 ** 2 ** 2 ** 2 ** 2",
    "( 1 - 2 ) / ( 3 - 5 )",
    "9 / 8 * 8",
    "64 / ( 8 * 8 )",
]

for expression in expressions:
    print(evaluate(expression))
```
